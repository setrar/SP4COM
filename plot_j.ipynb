```julia
using Plots
```

```julia
j = âˆšComplex(-1)
```

```julia
# Define the points
points = [j, -j];
```

```julia
# Add grid
p = plot(
    grid=:on
    , gridstyle=:dash
    , gridcolor=:gray
    , size=(600, 600)
    , legend=false
)


# Plot
p = plot!(
    xlims=(-2, 2), ylims=(-2, 2)
    , title="Plot of j and -j on the Complex Plane"
    , xlabel="Real part", ylabel="Imaginary part"
    , xticks=-2:1:2, yticks=-2:1:2
)

# Plot points as red circles
p = scatter!(
    real.(points), imag.(points)
    , color=:red, shape=:circle
    , label="Points"
)


# Add zero lines for reference
p = hline!([0]
    , color=:black
    , linestyle=:solid, linewidth=0.5
)
p = vline!([0]
    , color=:black
    , linestyle=:solid, linewidth=0.5
)

# Add annotations
p = annotate!(
    real.(points)[1], imag.(points)[1]+0.2
    , text("j", :left, 10)
)
p = annotate!(
    real.(points)[2], imag.(points)[2]-0.2
    , text("-j", :left, 10)
)

# Display the plot
display(p)
```
