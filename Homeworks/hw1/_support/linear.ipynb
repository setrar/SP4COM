To transform the given linear function $ y = \beta_0 + \beta_1 \times x $ into matrix notation using $ W $ as the coefficient matrix, follow these steps:

1. **Represent the coefficients and variables as vectors or matrices:**
   - Let $ W $ be the coefficient matrix or vector, containing $ \beta_0 $ and $ \beta_1 $.
   - Let $ X $ be the input vector, where we augment $ x $ with a constant 1 to accommodate the intercept $ \beta_0 $.

2. **Define $ W $ and $ X $:**
   - $ W = \begin{bmatrix} \beta_0 \\ \beta_1 \end{bmatrix} $
   - $ X = \begin{bmatrix} 1 \\ x \end{bmatrix} $

3. **Matrix multiplication for the linear model:**
   - The matrix product $ W^T X $ (transpose of $ W $ times $ X $) gives the desired linear equation in matrix form. Here, $ W^T $ is the transpose of $ W $, making it a row vector.

   $
   y = W^T X = \begin{bmatrix} \beta_0 & \beta_1 \end{bmatrix} \begin{bmatrix} 1 \\ x \end{bmatrix} = \beta_0 \times 1 + \beta_1 \times x
   $

   This results in the same equation as the original, $ y = \beta_0 + \beta_1 \times x $, but expressed in matrix notation using $ W $ as the coefficients.

By using this matrix formulation, you can extend the model to multiple linear regression more naturally, where $ X $ would include additional predictors, and $ W $ would contain their corresponding coefficients.

```julia
W = [ β₀; β₁]; X = [1 ; x];
```

```julia
y₁ = transpose(W) .* X;
```

```julia
# Create the plot
pl1 = plot(y₁
    , label=L" y = β_0 + β_1 x"
    , title="Graph of the line " * L" y = W^T X"
    , xlabel="x", ylabel="y"
    , margin = 5mm
)
```
