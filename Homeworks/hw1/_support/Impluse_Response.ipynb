```julia
using Plots
using DSP
```

```julia


# Define a simple discrete signal (e.g., a step function)
n = 0:20
signal = [ones(10); zeros(11)]  # A step that goes to zero after 10 samples

# Define the impulse response of a moving average filter
h = ones(5) / 5  # A simple average of 5 samples

# Convolve the signal with the impulse response to get the output
output = conv(signal, h)

# Plot the original signal
plot(n, signal, label="Original Signal", title="Signal and Its Impulse Response", xlabel="Sample", ylabel="Amplitude")

# Plot the impulse response (shift appropriately for visualization)
h_n = 0:length(h)-1
plot!(h_n, h, label="Impulse Response")

# Extend n to match the length of the convolution output
n_extended = 0:length(output)-1

# Plot the filtered signal (output of convolution)
plot!(n_extended, output, label="Filtered Signal (Output)")

# Display the plot
# display(plot)
```
