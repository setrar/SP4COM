```julia
using DSP, Plots
```

```julia
# Define the sample rate and duration
fs = 1000  # Sample rate in Hz
t = 0:1/fs:1.0  # Time vector for 1 second

# Create a test signal: sum of a low-frequency and high-frequency sine waves
signal = sin.(2π * 10 * t) + 0.5 * sin.(2π * 200 * t);
```

```julia
# Design a low-pass FIR filter using a Hamming window
cutoff_frequency = 50  # Cutoff frequency in Hz
num_taps = 51  # Number of taps in the filter
filter = digitalfilter(
    Lowpass(
        cutoff_frequency/(fs/2), fs=fs)
        , FIRWindow(hamming(num_taps)
    )
);
```

```julia
# Apply the FIR filter to the signal
filtered_signal = filt(filter, signal)

# Plotting the original and filtered signals
plot(t, signal, label="Original Signal", title="FIR Filter Example", xlabel="Time (seconds)", ylabel="Amplitude")
plot!(t, filtered_signal, label="Filtered Signal", linewidth=2)
```
