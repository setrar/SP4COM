```julia
using Plots
using FFTW

# Generate a synthetic time-domain signal: a sine wave
Fs = 1000  # Sampling frequency (Hz)
t = 0:1/Fs:1-1/Fs  # Time vector
f0 = 5  # Frequency of the sine wave (Hz)
signal = sin.(2 * f0 * t)

# Compute the Fourier Transform of the signal
fft_result = fft(signal)

# Calculate the Power Spectral Density (PSD)
psd = abs.(fft_result).^2 / length(signal)

# Frequency vector (only half is needed because of symmetry)
freq = (0:length(signal)2-1) * Fs / length(signal)

# Plotting
plot(freq, psd[1:length(freq)], label="PSD", xlabel="Frequency (Hz)", ylabel="Power Density", title="Power Spectral Density")
scatter!(freq, real.(fft_result[1:length(freq)]).^2, label="Power of Real Part", marker=:circle)
scatter!(freq, imag.(fft_result[1:length(freq)]).^2, label="Power of Imaginary Part", marker=:square)
```
