```julia
using LinearAlgebra
using Plots
using FFTW
```

```julia
include("_support/operations.jl")
```

```julia
# Define the matrix A (suitable for 3D visualization)
# A = [1 0; 0 1; 1 1]  # Example matrix with two independent columns
A = [1 1; 1 2; 1 3]  # Example matrix with two independent columns

# Define the vector b
# b = [2; 2; 2];  # Example vector
b = [1; 3; 2];  # Example vector
```

```julia
# Compute the projection matrix that projects onto the column space of A
# Use the formula P = A (Aᵀ A)⁻¹ Aᵀ
Pₐ = A * ((A)ᵀ * A)⁻¹ * (A)ᵀ

# Calculate the projection of b onto the column space of A
Pᵦ₋ₐ = Pₐ * b;; @show Pᵦ₋ₐ
```

```julia

plot(title="Vector Projection onto Column Space of A"
    , xlabel="X", ylabel="Y", zlabel="Z"
    , legend=:topright
)

# Plot the original vector b
plot!([0, b[1]], [0, b[2]], [0, b[3]]
    , color=:blue, label="Original Vector b", linewidth=2
)

# Plot the projection of b
plot!([0, Pᵦ₋ₐ[1]], [0, Pᵦ₋ₐ[2]], [0, Pᵦ₋ₐ[3]]
    , color=:green, label="Projection of b", linewidth=2
)

# Optionally, plot the column vectors of A to provide context
quiver!([0, 0], [0, 0], [0, 0]
    , quiver=([A[1, 1], A[1, 2]]
    , [A[2, 1], A[2, 2]], [A[3, 1], A[3, 2]])
    , color=[:red  :red]
    , label=["Column 1 of A"  "Column 2 of A"], linewidth=2
)
```
