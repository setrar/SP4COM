```julia
using Luxor

function draw_diagram()
    Drawing(800, 600, "ICMF_diagram.png") # Creates a canvas of 800x600 pixels
    origin() # Center the drawing origin
    
    # Define boxes
    boxes = Dict(
        "Input y[k]" => (Point(-300, -150), 150, 100),
        "Spatial Filter h^H" => (Point(-150, 0), 150, 100),
        "d[k]" => (Point(0, -150), 150, 100),
        "Orthogonal Filter h^⊥H" => (Point(-150, 150), 150, 100),
        "x[k]" => (Point(0, 150), 150, 100),
        "Adaptive Filter f" => (Point(150, 150), 150, 100),
        "ε[k]" => (Point(300, 0), 150, 100)
    )

    # Draw boxes
    for (label, (pos, w, h)) in boxes
        sethue("white")
        box(pos, w, h, :fill)
        sethue("black")
        text(label, pos + Point(0, 5), halign=:center)
    end

    # Draw arrows
    arrow(Point(-225, -75), Point(-75, -75), linewidth=2)
    arrow(Point(-75, 0), Point(75, 0), linewidth=2)
    arrow(Point(-225, 75), Point(-75, 75), linewidth=2)
    arrow(Point(75, -75), Point(225, 0), linewidth=2)
    arrow(Point(75, 75), Point(225, 75), linewidth=2)
    arrow(Point(225, 75), Point(300, 15), linewidth=2)

    # Label arrows
    text("a[k], v[k]", Point(-225, -90), halign=:center)
    text("y[k]", Point(-75, 15), halign=:center)
    text("y[k]", Point(-75, 165), halign=:center)
    text("ε[k](f)", Point(225, 90), halign=:center)

    finish() # Finalize the drawing
    preview() # Open the image in a default viewer
end

draw_diagram()
```
