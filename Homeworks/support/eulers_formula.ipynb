```julia
# Add grid
p = plot(
    grid=:on
    , gridstyle=:dash
    , gridcolor=:gray
    , size=(600, 600)
    , legend=false
)


# Plot
p = plot!(
    xlims=(-2, 2), ylims=(-2, 2)
    , title="Plot of j and -j on the Complex Plane"
    , xlabel="Real part", ylabel="Imaginary part"
    , xticks=-2:1:2, yticks=-2:1:2
)

# Plot points as red circles
p = scatter!(
    real.(points), imag.(points)
    , color=:red, shape=:circle
    , label="Points"
)


# Add zero lines for reference
p = hline!([0]
    , color=:black
    , linestyle=:solid, linewidth=0.5
)
p = vline!([0]
    , color=:black
    , linestyle=:solid, linewidth=0.5
)

# Add annotations
p = annotate!(
    real.(points)[1], imag.(points)[1]+0.2
    , text("j", :left, 10)
)
p = annotate!(
    real.(points)[2], imag.(points)[2]-0.2
    , text("-j", :left, 10)
)
```

### why are complex numbers used in digital signal processing?

Complex numbers are widely used in digital signal processing (DSP) for several reasons that make them extremely useful in analyzing and processing signals, particularly those that vary over time or space. Here are the main reasons:

1. **Representation of Sinusoids**: Complex numbers provide a concise way to represent sinusoidal functions, which are fundamental to understanding signals in the time and frequency domains. The Euler's formula, $e^{j\theta} = \cos(\theta) + j\sin(\theta)$, where $j$ is the square root of -1, allows sinusoids to be represented as complex exponentials. This representation simplifies the analysis and manipulation of sinusoids, making operations like differentiation, integration, and multiplication by a constant easier.

2. **Frequency Domain Analysis**: Many DSP techniques involve transforming signals from the time domain to the frequency domain, where the representation and manipulation of signals become more intuitive. The Fourier Transform, which is central to frequency domain analysis, uses complex numbers to decompose a signal into its constituent frequencies. Complex numbers here facilitate the understanding of phase and amplitude of different frequency components of a signal.

3. **Signal Modulation and Demodulation**: Complex numbers are used in modulation schemes, where a signal is encoded for transmission over a medium. Quadrature Amplitude Modulation (QAM) and Phase Shift Keying (PSK) are examples where complex numbers represent the amplitude and phase of a carrier signal, allowing for efficient data transmission.

4. **Filter Design and Implementation**: In DSP, filters are used to enhance or suppress certain aspects of a signal. The design and analysis of filters, particularly in the frequency domain, often use complex numbers. For instance, the transfer function of a filter, which describes how different frequencies are attenuated or passed, is expressed in terms of complex frequencies.

5. **Signal Analysis Tools**: Tools like the Z-transform, which extends the discrete Fourier transform (DFT) by incorporating the concept of poles and zeros in the complex plane, rely on complex numbers. The Z-transform is crucial for analyzing the stability and frequency response of digital filters and systems.

6. **Efficiency in Computation**: Using complex numbers can lead to more efficient algorithms for signal processing. For example, the Fast Fourier Transform (FFT), an algorithm for computing the DFT of a signal efficiently, leverages the properties of complex numbers to significantly reduce the number of computations required, thus speeding up the processing of signals.

7. **Phase Information**: Complex numbers inherently capture both magnitude and phase information of signals. This is important in many DSP applications where phase shifts need to be accurately tracked and corrected, such as in communication systems and radar signal processing.

In summary, complex numbers are indispensable in DSP due to their ability to simplify the representation and manipulation of sinusoidal signals, enhance the efficiency of signal analysis and processing algorithms, and accurately capture the amplitude and phase information of signals.
