```julia
using Random, Distributions, StatsPlots, Plots;
gr();
```

```julia
Random.seed!(0)
μ, n = 20, 1000;
```

```julia
dist₁, dist₂ = rand(Normal(μ,μ),n), rand(Exponential(μ),n);
```

```julia
qqnorm(dist₁
    , color=:blue
    , ms=3, msw=0
    , label="Normal Data"
)
qqnorm!(dist₂
   , color=:red
    , ms=3, msw=0
    , label="Exponential Data"
    , xlabel="Normal Theoretical Quantiles", ylabel="Data Quantiles"
    , legend=true
)
```
