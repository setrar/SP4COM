```julia
using Plots, Plots.Measures, LaTeXStrings
```

```julia
# Define the coefficients
β₀, β₁ = 3, 2; @show β₀, β₁;
```

```julia
# Create an array of x values from -10 to 10
x = range(-10, stop=10, length=400); @show x;
```

```julia
# Calculate the corresponding y values
y = β₀ .+ β₁ * x;
```

```julia
# Create the plot
pl1 = plot(x, y
    , label=L" y = β_0 + β_1 x"
    , title="Graph of the line " * L" y = β_0 + β_1 x"
    , xlabel="x", ylabel="y"
    , margin = 5mm
)

# Add a horizontal line at y = 0
hline!(pl1, [0], color=:black, linewidth=0.5, label = :false)

# Add a vertical line at x = 0
vline!(pl1, [0], color=:black, linewidth=0.5, label = :false)

# Add grid and make final adjustments
plot!(pl1; grid=true, linestyle=:dash, color="gray", linewidth=0.5, legend=:true)
```
